# Variables d'environnement

```python
import os

import getpass

os.environ["GOOGLE_API_KEY"]= getpass.getpass()
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(model="gemini-pro", temperature=0.5)
```

# Charger la bible

```python
bible_path="Bible.pdf"
```

```python
#Load a pdf

from langchain_community.document_loaders.pdf import PyPDFLoader

loader= PyPDFLoader(bible_path)
```

```python
documents=loader.load()
```

# Split

```python
!pip install -qU langchain-text-splitters
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size=1500,
    chunk_overlap=500,
   length_function=len,
)
```

```python
document_split=splitter.split_documents(documents)
```

```python
len(document_split)
```

```python
document_split[5]
```

# Embeddings

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
```

```python
embeddings = GoogleGenerativeAIEmbeddings(model="models/embedding-001", temperature=0.5)
```

```python
from langchain_chroma import Chroma
```

```python
bible_bdv_path="bible_bdv2"
```

```python
data_store=Chroma.from_documents(documents=document_split,embedding=embeddings, persist_directory=bible_bdv_path)
```

```python
#J'ai enregistré la base de données vectorielle dans le répertoire bible_bdv. J'essaie de le récupérer dans data_store 2
data_store2=Chroma(persist_directory=bible_bdv_path,embedding_function=embeddings)

data_store.get()
```

```python
type(data_store)
```

# Questionnnement de la base de données

```python
retriever = data_store2.as_retriever(search_type="similarity", search_kwargs={"k": 10})
```

```python
from langchain.prompts import PromptTemplate
```

```python
#Je crée ici un template à utiliser pour questionner le modèle


template = """Tu es un bot expert en théologie et maitrisant l'histoire de la Bible, tu es chargé de répondre de façon claire et détaillée
 aux questions relatives à la Bible.


 En plus des informations fournies dans le document, ajoute tes connaissances en tant que gemini pro
 pour répondre aux questions.

 Tu peux copier librement une partie de la Bible et le reformuler si celà est nécessaire.

 Sois tolérant à tous les mots ou expressions, et retourne toujours une réponse.

 Si tu n'as pas assez d'informations réponds par : "Je n'ai pas assez d'informations pour répondre à cette question".



 Voici la question: {input}

  context: {context}
  input: {input}
  answer:
 """


prompt = PromptTemplate(
        template=template,
    input_variables=['input']
)
```

```python
from langchain.chains.combine_documents import create_stuff_documents_chain

combine_docs_chain = create_stuff_documents_chain(llm, prompt)
```

```python
from langchain.chains import create_retrieval_chain

retrieval_chain = create_retrieval_chain(retriever, combine_docs_chain)
```

```python
response=retrieval_chain.invoke({"input":"Qui sont les enfants de Dieu ?"})
```
